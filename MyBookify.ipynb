```python
import spotipy
import json
import webbrowser
import urllib.request
import spotipy.util as util
import sys
```

```python
# copy and paste the url into your browser to see what data you are getting back
request = urllib.request.Request("https://www.googleapis.com/books/v1/volumes?q=search+terms")
response =urllib.request.urlopen(request)
```

```python
type(response)
```

```python
books = json.loads(response.read())
```

```python
print(books)
```

```python
subject = books["items"][0]["volumeInfo"]["description"]
categories = books["items"][0]["volumeInfo"]["categories"]
```

```python
subject
categories
```

```python
credentials = "Spotipy_keys.json"
with open(credentials, "r") as keys:
    api_tokens = json.load(keys)
```

```python
client_id = api_tokens["client_id"]
client_secret = api_tokens["client_secret"]
redirectURI = api_tokens["redirect"]
username = api_tokens["username"]
```

```python
scope = 'user-read-private user-read-playback-state user-modify-playback-state playlist-modify-public'
token = util.prompt_for_user_token(username, scope, client_id=client_id,
                           client_secret=client_secret,
                           redirect_uri=redirectURI)
```

```python
sp = spotipy.Spotify(auth=token)
```

```python
track_results = sp.search(q=categories, type='track', limit=25)
```

```python
tracks = track_results['tracks']['items']
```

```python
tracks[0]
```

```python
track_selection_list = []
track_art = []

for song in tracks:
    track_selection_list.append(song['uri'])
    track_art.append(song['album']['images'][0]['url'])
```

```python
print(username)
print(categories[0])
```

```python
my_playlist = sp.user_playlist_create(user=username, name=categories[0], public=True,
                                      description="Songs for the bookLover")
results = sp.user_playlist_add_tracks(username, my_playlist['id'], track_selection_list)
print(results)
print(my_playlist)
```
